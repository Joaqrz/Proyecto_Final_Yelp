```python
import pandas as pd     
import seaborn as sns
import os
```

```python
rutaJson = "D:/PROYECTO FINAL/Google Maps/metadata-sitios/"     # Ruta de los json de Sitios
```

```python
archivos_json = [os.path.join(rutaJson, archivo) for archivo in os.listdir(rutaJson) if archivo.endswith('.json')]
```

```python
archivos_json
```

```python
# Cargar todas los archivos Json y colocarlos en una lista
lista_df_sitios = []
for archivo in archivos_json:       
    df_sitio = pd.read_json(archivo, lines=True)
    lista_df_sitios.append(df_sitio)
```

```python
df_sitios = pd.concat(lista_df_sitios, ignore_index=True)     # Concatenar los Dataframes cargados en la lista
```

```python
categorias = df_sitios['category'].explode().unique()    # Cargar las categorias en arreglo
df_categorias = pd.DataFrame({'category':categorias})   # Cargar las categorias en un df
df_categorias['Id'] = range(1, len(df_categorias) + 1)  # Crear un Id correlativo para cada categoria
df_categorias = df_categorias[['Id','category']]        # Camabiar el orden de las columnas
```

```python
df_sitios.to_parquet('D:/PROYECTO FINAL/Google Maps/Sitios_GMaps.parquet', index=False)
```
