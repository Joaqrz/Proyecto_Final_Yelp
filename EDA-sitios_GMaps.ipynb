### **2. Analisis de Datos Categoricos**

```python
df_sitios['price'].value_counts()
```

```python
# Grafica el countplot
sns.countplot(y='category', data=df_category, order=df_category['category'].value_counts().iloc[:10].index)
plt.ylabel('Categoria')
plt.xlabel('Cantidad')
plt.title('Distribucion de Sitios por Categoria')
plt.show()
```

```python
# Grafica el countplot
sns.countplot(y='service_option', data=df_Service_options, order=df_Service_options['service_option'].value_counts().iloc[:10].index)
plt.ylabel('Categoria')
plt.xlabel('Cantidad')
plt.title('Distribucion de Sitios por Opcion de Servicio')
plt.show()
```

```python
# Grafica el countplot
sns.countplot(y='planning_option', data=df_Planning_options, order=df_Planning_options['planning_option'].value_counts().iloc[:10].index)
plt.ylabel('Categoria')
plt.xlabel('Cantidad')
plt.title('Distribucion de Sitios por Opcion de Planning')
plt.show()
```

```python
df_union_cat_sitios= pd.merge(df_category, df_sitios, on='gmap_id', how= 'inner')
df_grupo_estados= df_union_cat_sitios.groupby('estado', as_index=False, observed=False)['category'].value_counts().sort_values(by='count', ascending=False)
df_grupo_estados= df_grupo_estados[(df_grupo_estados['category']=="Restaurant") & (df_grupo_estados['estado'] != "SIN DATO")]
```

```python
plt.figure(figsize=(6, 14))
sns.barplot(data= df_grupo_estados, y='estado', x='count')
plt.title("Cantidad de Restaurantes por Estado")
plt.xlabel("cantidad")
plt.ylabel("Estado")
plt.show()
```
