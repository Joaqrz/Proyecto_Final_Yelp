#FUNCION CARGA JSONS

```python
# cargar_json_a_dataframe()

# Definición de la función para cargar archivos JSON en un DataFrame
def cargar_json_a_dataframe(ruta):
    # Abre el archivo JSON y lee las líneas
    with open(ruta, 'r', encoding='utf-8') as f:
        data = f.readlines()

    # Convierte cada línea en un diccionario usando json.loads
    registros = [json.loads(line.strip()) for line in data]

    # Crea el DataFrame a partir de la lista de diccionarios
    df = pd.DataFrame(registros)

    return df


# EJEMPLO Para llamar funcion 
#         ruta = '../../Google Maps/Revies- States/review-Alabama/1.json'

#         reviews_1_alabama = cargar_json_a_dataframe(ruta)
#         reviews_1_alabama
```
